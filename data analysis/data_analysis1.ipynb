```python
Hispanic_students = student_data[student_data["Hispanic"] ==1]
AmericanIndian_students= student_data[student_data["AmericanIndian"] == 1]
Asian_students = student_data[student_data["Asian"] == 1]
Black_student = student_data[student_data["Black"] == 1]
NativeHawaiian_student = student_data[student_data["NativeHawaiian"] == 1]
White_students = student_data[student_data["White"] == 1]
TwoOrMoreRaces_students = student_data[student_data["TwoOrMoreRace"] == 1]
```

```python
Hispanic_students_dropout = student_data_dropout[student_data_dropout["Hispanic"] ==1]
AmericanIndian_students_dropout = student_data_dropout[student_data_dropout["AmericanIndian"] == 1]
Asian_students_dropout = student_data_dropout[student_data_dropout["Asian"] == 1]
Black_student_dropout = student_data_dropout[student_data_dropout["Black"] == 1]
NativeHawaiian_student_dropout = student_data_dropout[student_data_dropout["NativeHawaiian"] == 1]
White_students_dropout = student_data_dropout[student_data_dropout["White"] == 1]
TwoOrMoreRaces_students_dropout = student_data_dropout[student_data_dropout["TwoOrMoreRace"] == 1]
```

```python
#plot of dropouts based on nationality

#define data
nationality_dropout_data = [len(Hispanic_students_dropout), len(AmericanIndian_students_dropout) , len(Asian_students_dropout), len(Black_student_dropout), len(NativeHawaiian_student_dropout), len(White_students_dropout), len(TwoOrMoreRaces_students_dropout)]
nationality_dropout_labels = ['Hispanic', 'American Indian', 'Asian', 'Black', 'Native Hawaiian', 'White', 'Two or More Races']

colors = sns.color_palette('pastel')[0:7]

plt.pie(nationality_dropout_data, labels = nationality_dropout_labels, colors = colors, autopct='%.0f%%')
plt.show()
```

```python
Hispanic_percentage = len(Hispanic_students_dropout) / len(Hispanic_students) 
Asian_percentage = len(Asian_students_dropout) / len(Asian_students)
Black_percentage = len(Black_student_dropout) / len(Black_student)
AmericanIndian_percentage = len(AmericanIndian_students_dropout) / len(AmericanIndian_students)
NativeHawaiian_percentage = len(NativeHawaiian_student_dropout) / len(NativeHawaiian_student)
White_percentage = len(White_students_dropout) / len(White_students)
TwoOrMoreRaces_percentage = len(TwoOrMoreRaces_students_dropout) / len(TwoOrMoreRaces_students)

percentage_data = {
    "name" : ["Hispanic", "Asian", "Black", "American Indian", "Native Hawaiian", "White", "Two or More Races"],
    "percentage" : [Hispanic_percentage, Asian_percentage, Black_percentage, AmericanIndian_percentage, NativeHawaiian_percentage, White_percentage, TwoOrMoreRaces_percentage]
}

percenbtage_df = pd.DataFrame(percentage_data)
```

```python
sns.barplot(x="name", y="percentage", data= percenbtage_df)
```
