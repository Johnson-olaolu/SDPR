```python
#student progress fall data 
student_progress_2011_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2011_SP.csv")
student_progress_2012_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2012_SP.csv")
student_progress_2013_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2013_SP.csv")
student_progress_2014_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2014_SP.csv")
student_progress_2015_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2015_SP.csv")
student_progress_2016_Fall = pd.read_csv("./datasets/Student Progress Data/Fall 2016_SP.csv")

#student progress spring data
student_progress_2012_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2012_SP.csv")
student_progress_2013_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2013_SP.csv")
student_progress_2014_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2014_SP.csv")
student_progress_2015_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2015_SP.csv")
student_progress_2016_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2016_SP.csv")
student_progress_2017_Spring = pd.read_csv("./datasets/Student Progress Data/Spring 2017_SP.csv")
```

```python
student_progress_2012_Spring.head()
```

```python
student_progress_Fall_data_array = [student_progress_2011_Fall, student_progress_2012_Fall, student_progress_2013_Fall, student_progress_2014_Fall, student_progress_2015_Fall, student_progress_2016_Fall]
student_progress_Spring_data_array = [student_progress_2017_Spring, student_progress_2012_Spring, student_progress_2013_Spring, student_progress_2014_Spring, student_progress_2015_Spring, student_progress_2016_Spring]
```

```python
for i in range(6):
    student_progress_Fall_data_array[i]['Season'] = "Fall"
    student_progress_Spring_data_array[i]['Season'] = "Spring"
student_progress_Fall_data_array[1].head()
```

```python
student_progress_Fall_data = pd.concat((student_progress_Fall_data_array[1], student_progress_Fall_data_array[2], student_progress_Fall_data_array[3], student_progress_Fall_data_array[4], student_progress_Fall_data_array[5], student_progress_Fall_data_array[0]))

student_progress_Spring_data = pd.concat((student_progress_Spring_data_array[0],student_progress_Spring_data_array[1], student_progress_Spring_data_array[2], student_progress_Spring_data_array[3], student_progress_Spring_data_array[4], student_progress_Spring_data_array[5]))
student_progress_Spring_data
```

```python
student_progress_data  = pd.concat((student_progress_Fall_data, student_progress_Spring_data))
```
